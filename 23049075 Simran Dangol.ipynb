# Statistical Testing

## Test 1: Whether the average response time across complaint types is similar or not.

### Null Hypothesis(H0): The average Request_Closing_Hours across complaint types is same.
### Alternate Hypothesis(H1): The average Request_Closing_Hours across complaint types is different.

```python
#Import the one-way f-test function from scipy.stats
from scipy.stats import f_oneway

#Select the top 5 most common complaint types
top_5_complaints= df['Complaint Type'].value_counts().head(5).index

#Filter the DataFrame to only include rows with those top 5 complaint types
filtered = df[df['Complaint Type'].isin(top_5_complaints)]

#Group the data by 'Complaint Type' and extract the 'Request_Closing_Hours' values
groups = [group['Request_Closing_Hours'] for name, group in filtered.groupby('Complaint Type')]

#Perform a F-test on the response time groups
f_stat, p_value = f_oneway(*groups)

#Display the test statistic and p-value
print("F-Test:")
print("F-statistic:", f_stat)
print("P-value:", p_value)

#Interpret the p-value to accept or reject the Null Hypothesis
if p_value < 0.05:
    print("Reject the null hypothesis: Response time differs by complaint type.")
else:
    print("Fail to reject the null hypothesis: No significant difference in response times.")
```

## Test 2: Whether the type of complaint or service requested and location are related.

### Null Hypothesis(H0): The complaint type and the location type are not related. 
### Alternate Hypothesis(H1): The complaint type and the location type are related. 

```python
#Import required chi-square test function from scipy.stats
from scipy.stats import chi2_contingency

#Create a contingency table for Complaint Type vs. Location Type
data = pd.crosstab(df['Complaint Type'], df['Location Type'])

#Perform the chi-square test 
chi2, p_value, dof, expected = chi2_contingency(data)

#Display the test statistic and p-value
print("Chi-Square Test:")
print("Chi2 Statistic:", chi2) 
print("P-value:", p_value)

#Interpret the p-value to accept or reject the Null Hypothesis
if p_value < 0.05:
    print("Reject the null hypothesis: Complaint Type and Location Type are related.")
else:
    print("Fail to reject the null hypothesis: No relationship between Complaint Type and Location Type.")
```
